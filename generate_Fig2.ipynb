```python
# --- Config ---
CANVAS_H, CANVAS_W = 10473, 14735  # (H, W)
# List your images with their top-left insertion coordinates (y, x)
TILES = [
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q005.ome.tiff", "y": 0, "x": 4130},
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q006.ome.tiff", "y": 743, "x": 0},
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q007.ome.tiff", "y": 4925, "x": 4076},
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q008.ome.tiff", "y": 4925+354, "x": 4076},
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q009.ome.tiff", "y": 1437, "x": 8398},
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q010.ome.tiff", "y": 4916, "x": 10565},
    {"path": "output/20250714_EMMET_20250714-160145-670957_Q011.ome.tiff", "y": 3514, "x": 4088},
]

# --- Helpers ---
def to_chw_uint8(arr: np.ndarray) -> np.ndarray:
    """
    Convert an image array to channel-first (C,H,W) uint8.
    Handles HWC/CHW, RGB/RGBA, uint8/uint16/float.
    """
    if arr.ndim != 3:
        raise ValueError(f"Expected 3D image, got shape {arr.shape}")

    # Bring to RGB and channel-first
    if arr.shape[-1] in (3, 4):         # HWC
        arr = arr[..., :3]              # drop alpha if present
        chw = arr.transpose(2, 0, 1)    # (C,H,W)
    elif arr.shape[0] in (3, 4):        # CHW
        chw = arr[:3]                   # drop alpha if present
    else:
        raise ValueError(f"Can't infer channels from shape {arr.shape}")

    # Convert to uint8 if needed
    if chw.dtype == np.uint8:
        return chw

    arr_f = chw.astype(np.float32)
    maxv = arr_f.max()
    minv = arr_f.min()
    if maxv == minv:
        return np.zeros_like(chw, dtype=np.uint8)
    arr_n = (arr_f - minv) / (maxv - minv) * 255.0
    return np.clip(arr_n, 0, 255).astype(np.uint8)

def paste_tile(canvas: np.ndarray, tile: np.ndarray, y0: int, x0: int) -> None:
    """
    Paste (C,H,W) uint8 tile onto the (C,H,W) uint8 canvas at (y0, x0).
    Tiles that extend beyond canvas bounds are cropped.
    """
    Cc, Hc, Wc = canvas.shape
    Ct, Ht, Wt = tile.shape
    if Ct != 3 or Cc != 3:
        raise ValueError("Expecting 3 channels (RGB) for both canvas and tiles.")

    # Compute fit
    if y0 >= Hc or x0 >= Wc:
        return  # entirely outside; nothing to paste

    h_fit = max(0, min(Ht, Hc - y0))
    w_fit = max(0, min(Wt, Wc - x0))
    if h_fit <= 0 or w_fit <= 0:
        return

    canvas[:, y0:y0 + h_fit, x0:x0 + w_fit] = tile[:, :h_fit, :w_fit]

# --- 1) Create a WHITE color canvas (C,H,W) ---
canvas = np.full((3, CANVAS_H, CANVAS_W), 255, dtype=np.uint8)

# --- 2) Load each OME-TIFF and paste it ---
for i, item in enumerate(TILES, start=1):
    p = Path(item["path"])
    y0, x0 = int(item["y"]), int(item["x"])
    img = tiff.imread(str(p))      # full-resolution read; tifffile handles OME-TIFF
    tile = to_chw_uint8(img)       # normalize & force (C,H,W) uint8
    paste_tile(canvas, tile, y0, x0)

# --- 3) Package as `result` and display ---
result = [canvas]  # so result[0] is (C,H,W), matching your display call

plt.figure(figsize=(10, 10))
plt.imshow(result[0].transpose(1, 2, 0))  # (H,W,C) for imshow
plt.axis('off')
plt.title('Brightfield Image')
plt.show()
```

```python
def crop_chw(img: np.ndarray, x: int, y: int, w: int, h: int) -> np.ndarray:
    """
    Crop (C,H,W) by pixel-rect (x,y,w,h). Clamps to bounds.
    x -> columns, y -> rows (origin top-left).
    """
    C, H, W = img.shape
    x0 = max(0, int(x))
    y0 = max(0, int(y))
    x1 = min(W, int(round(x + w)))
    y1 = min(H, int(round(y + h)))
    if x1 <= x0 or y1 <= y0:
        return np.zeros((C, 0, 0), dtype=img.dtype)
    return img[:, y0:y1, x0:x1]

def resize_nearest_chw(tile: np.ndarray, scale: float) -> np.ndarray:
    """
    Nearest-neighbour resize for (C,H,W) to arbitrary scale (>0).
    No extra dependencies.
    """
    assert scale > 0, "scale must be > 0"
    C, H, W = tile.shape
    Hn = max(1, int(round(H * scale)))
    Wn = max(1, int(round(W * scale)))
    yi = np.floor(np.arange(Hn) / scale).astype(int)
    xi = np.floor(np.arange(Wn) / scale).astype(int)
    yi = np.clip(yi, 0, H - 1)
    xi = np.clip(xi, 0, W - 1)
    # (C,Hn,Wn) via advanced indexing/broadcast
    return tile[:, yi[:, None], xi[None, :]]

def paste_with_alpha(canvas: np.ndarray, tile: np.ndarray, y0: int, x0: int, alpha: float = 1.0):
    """
    Paste (C,H,W) tile onto (C,H,W) canvas at (y0,x0) with optional alpha in [0,1].
    Returns (h_fit, w_fit) actually pasted (after clipping). (0,0) if nothing pasted.
    """
    Cc, Hc, Wc = canvas.shape
    Ct, Ht, Wt = tile.shape
    if Ct != 3 or Cc != 3:
        raise ValueError("Expect 3-channel RGB.")
    if y0 >= Hc or x0 >= Wc:
        return 0, 0

    h_fit = max(0, min(Ht, Hc - y0))
    w_fit = max(0, min(Wt, Wc - x0))
    if h_fit <= 0 or w_fit <= 0:
        return 0, 0

    if alpha >= 1.0:
        canvas[:, y0:y0+h_fit, x0:x0+w_fit] = tile[:, :h_fit, :w_fit]
    else:
        dst = canvas[:, y0:y0+h_fit, x0:x0+w_fit].astype(np.float32)
        src = tile[:, :h_fit, :w_fit].astype(np.float32)
        out = (1.0 - alpha) * dst + alpha * src
        canvas[:, y0:y0+h_fit, x0:x0+w_fit] = np.clip(out, 0, 255).astype(np.uint8)
    return h_fit, w_fit

def draw_rect_chw(img: np.ndarray, x: int, y: int, w: int, h: int,
                  color=(0, 0, 0), thickness: int = 3) -> None:
    """
    Draw an axis-aligned rectangle on (C,H,W) in-place. Default color: black.
    """
    C, H, W = img.shape
    x0 = max(0, int(x)); y0 = max(0, int(y))
    x1 = min(W, int(round(x + w))); y1 = min(H, int(round(y + h)))
    if x1 <= x0 or y1 <= y0:
        return
    col = np.array(color, dtype=np.uint8)[:, None, None]
    t = max(1, int(thickness))
    # top/bottom
    img[:, y0:y0+t, x0:x1] = col
    img[:, max(y1-t, 0):y1, x0:x1] = col
    # left/right
    img[:, y0:y1, x0:x0+t] = col
    img[:, y0:y1, max(x1-t, 0):x1] = col

def draw_line_chw(img: np.ndarray, x0: int, y0: int, x1: int, y1: int,
                  color=(0, 0, 0), thickness: int = 2) -> None:
    """
    Draw a line on (C,H,W) using simple sampling + square brush.
    """
    _, H, W = img.shape
    x0 = float(x0); y0 = float(y0)
    x1 = float(x1); y1 = float(y1)
    steps = int(max(abs(x1 - x0), abs(y1 - y0))) + 1
    if steps <= 0:
        return
    xs = np.linspace(x0, x1, steps)
    ys = np.linspace(y0, y1, steps)
    r = max(0, int(np.ceil(thickness / 2)))
    col = np.array(color, dtype=np.uint8)[:, None, None]
    for xf, yf in zip(xs, ys):
        cx = int(round(xf))
        cy = int(round(yf))
        xa, xb = max(0, cx - r), min(W, cx + r + 1)
        ya, yb = max(0, cy - r), min(H, cy + r + 1)
        if xa < xb and ya < yb:
            img[:, ya:yb, xa:xb] = col

def compute_paste_fit(canvas_shape, tile_shape, y0: int, x0: int):
    """
    Return (h_fit, w_fit) describing how much of `tile_shape` will fit on canvas at (y0,x0).
    Shapes are (C,H,W).
    """
    _, Hc, Wc = canvas_shape
    _, Ht, Wt = tile_shape
    if y0 >= Hc or x0 >= Wc:
        return 0, 0
    h_fit = max(0, min(Ht, Hc - y0))
    w_fit = max(0, min(Wt, Wc - x0))
    return h_fit, w_fit

def add_margins_chw(img: np.ndarray, top=0, bottom=0, left=0, right=0,
                    color=(255, 255, 255)) -> Tuple[np.ndarray, Tuple[int, int]]:
    """
    Pad a (C,H,W) image with solid color margins. Returns (padded, (off_x, off_y)).
    """
    C, H, W = img.shape
    Hn, Wn = H + top + bottom, W + left + right
    out = np.empty((C, Hn, Wn), dtype=img.dtype)
    out[:] = np.array(color, dtype=img.dtype)[:, None, None]  # fill
    out[:, top:top+H, left:left+W] = img
    return out, (left, top)

def annotate_inset(ax, meta: dict, title: str,
                   color=(0, 0, 0), where='above',
                   pad=30, swatch_w=40, swatch_h=18,
                   fontsize=14):
    """
    Draw a small color swatch + title centered above/below the inset.
    `meta` is the dict returned by overlay_zoomed_roi.
    """
    x, y, w, h = meta["dst_x"], meta["dst_y"], meta["w"], meta["h"]
    cx = x + w / 2.0
    ty = (y - pad) if where == 'above' else (y + h + pad)

    # Color swatch (with thin black edge)
    ax.add_patch(Rectangle(
        (cx - (swatch_w + 6) / 2.0, ty - swatch_h / 2.0),
        swatch_w, swatch_h,
        facecolor=(color[0]/255, color[1]/255, color[2]/255),
        edgecolor='black', linewidth=1, transform=ax.transData
    ))

    # Text next to swatch
    ax.text(cx + swatch_w/2.0 + 6, ty, title,
            ha='left', va='center', fontsize=fontsize, color='black',
            bbox=dict(facecolor='white', alpha=0.75, edgecolor='none', pad=2),
            transform=ax.transData)

def _to_mpl_color(c):
    """Accept (R,G,B) in 0–255 or a Matplotlib color string."""
    if isinstance(c, str):
        return c
    r, g, b = c
    return (r/255.0, g/255.0, b/255.0)

def annotate_inset_with_legend(
    ax,
    meta: dict,
    title: str,
    entries: List[Tuple[str, Union[Tuple[int, int, int], str]]],  # [(label, color), ...]
    *,
    where: str = "above",          # "above" or "below" relative to inset
    pad_title: int = 28,           # gap from inset to title
    title_fontsize: int = 16,
    legend_left: str = "inset",    # "inset" = align legend to inset's left edge
    legend_offset: int = 10,       # px from left edge of inset
    swatch_w: int = 28,
    swatch_h: int = 14,
    row_gap: int = 8,
    item_fontsize: int = 14,
    box_alpha: float = 0.85
):
    """
    Draw a centered title and a vertical legend block directly *below* the title.
    Coordinates are in image pixels (ax.transData).
    """
    x, y, w, h = meta["dst_x"], meta["dst_y"], meta["w"], meta["h"]
    cx = x + w / 2.0
    # Title position
    title_y = (y - pad_title) if where == "above" else (y + h + pad_title)

    # Title (centered)
    ax.text(
        cx, title_y, title,
        ha="center", va="center", fontsize=title_fontsize, color="black",
        bbox=dict(facecolor="white", alpha=box_alpha, edgecolor="none", pad=2),
        transform=ax.transData
    )

    # Legend block starts *below* the title in either case
    start_y = title_y + swatch_h + row_gap
    # Left edge for legend lines
    left_x = (x + legend_offset) if legend_left == "inset" else (cx - swatch_w/2 - 40)

    for i, (label, color) in enumerate(entries):
        yi = start_y + i * (swatch_h + row_gap)
        ax.add_patch(Rectangle(
            (left_x, yi - swatch_h/2), swatch_w, swatch_h,
            facecolor=_to_mpl_color(color), edgecolor="black", linewidth=1,
            transform=ax.transData
        ))
        ax.text(
            left_x + swatch_w + 8, yi, label,
            ha="left", va="center", fontsize=item_fontsize, color="black",
            bbox=dict(facecolor="white", alpha=box_alpha, edgecolor="none", pad=1),
            transform=ax.transData
        )

def overlay_zoomed_roi(
    base: np.ndarray, x: int, y: int, w: int, h: int,
    scale: float, dst_x: int, dst_y: int,
    *,
    alpha: float = 1.0,
    outline: bool = True,
    outline_color=(0, 0, 0),          # black
    outline_thickness: int = 20,
    zoom_border: bool = True,
    zoom_border_color=(0, 0, 0),      # black
    zoom_border_thickness: int = 20,
    connectors: bool = True,
    connector_color=(0, 0, 0),        # black
    connector_thickness: int = 6
) -> dict:
    """
    Draw order (ensures lines are behind the inset):
      1) connectors (behind)
      2) ROI outline
      3) paste zoomed inset
      4) border around inset (topmost)
    """
    # 1) Prepare ROI + zoomed tile
    roi = crop_chw(base, x, y, w, h)
    if roi.size == 0:
        raise ValueError("ROI is empty after clamping to bounds.")
    zoom = resize_nearest_chw(roi, scale)

    # 2) Compute clipped inset size for accurate connector endpoints
    fit_h, fit_w = compute_paste_fit(base.shape, zoom.shape, int(dst_y), int(dst_x))

    # 3) Draw connectors *first* so they are behind the inset
    if connectors and fit_h > 0 and fit_w > 0:
        src_corners = [
            (x,       y),         # TL
            (x + w,   y),         # TR
            (x,       y + h),     # BL
            (x + w,   y + h),     # BR
        ]
        dst_corners = [
            (int(dst_x),                 int(dst_y)),                 # TL
            (int(dst_x + fit_w - 1),     int(dst_y)),                 # TR
            (int(dst_x),                 int(dst_y + fit_h - 1)),     # BL
            (int(dst_x + fit_w - 1),     int(dst_y + fit_h - 1)),     # BR
        ]
        for (sx, sy), (dx, dy) in zip(src_corners, dst_corners):
            draw_line_chw(
                base, sx, sy, dx, dy,
                color=connector_color, thickness=connector_thickness
            )

    # 4) ROI outline (above connectors, below inset)
    if outline:
        draw_rect_chw(
            base, x, y, w, h,
            color=outline_color, thickness=outline_thickness
        )

    # 5) Paste inset (on top of lines)
    pasted_h, pasted_w = paste_with_alpha(
        base, zoom, int(dst_y), int(dst_x), alpha=alpha
    )

    # 6) Border around the inset (topmost)
    if zoom_border and pasted_h > 0 and pasted_w > 0:
        draw_rect_chw(
            base, int(dst_x), int(dst_y), int(pasted_w), int(pasted_h),
            color=zoom_border_color, thickness=zoom_border_thickness
        )

    return {
        "dst_x": int(dst_x),
        "dst_y": int(dst_y),
        "w": int(pasted_w),
        "h": int(pasted_h),
        "roi": (int(x), int(y), int(w), int(h)),
    }
```
